```python
import sys,os,time,jieba,gensim,re,math,codecs,random
from jieba import analyse
import pandas as pd
from tqdm import tqdm
mp=os.getcwd()
mp
```

```python
#本文所使用的数据
data=pd.read_excel(mp+'/data/all_data_with_singleKM.xlsx')
```

```python
#用户输入
inputwords=['高校图书馆']
```

```python
data['关键词相似度']=None
data['推荐指数']=None

from cos_dist import cos_dist

st=time.time()
for i in tqdm(range(len(data))):
    data['关键词相似度'].iloc[i]=cos_dist(list(set(data['关键词'].iloc[i].split(','))),inputwords)
#     data['被引频次'].iloc[i]=data['被引频次'].iloc[i]/50
    if data['关键词相似度'].iloc[i] != 0 :
        data['推荐指数'].iloc[i]=math.log(data['关键词相似度'].iloc[i] * (data['被引频次'].iloc[i]/(2020-data['年'].iloc[i])))
    else :
        data['推荐指数'].iloc[i]=0
et=time.time()
print('计算结束，共花费{}s'.format(et-st))
```

```python
#按推荐指数排列，取前n条
FinalZS=data[['年','题名','作者','摘要','abstractKM','推荐指数',]].sort_values(by=['推荐指数'],ascending=False).iloc[:30]

#按年排列
FinalZS=FinalZS.sort_values(by=['年'],ascending=True)
# FinalZS
```

```python
ZYlist=[]#要输入LDA训练的数据
for i in range(len(FinalZS)):
        ZYlist.append([FinalZS['摘要'].iloc[i]])

from LDA import genlda
kk=genlda(ZYlist,6)#聚类数量：6
label=[]
# print(label)
# 取出每一个的Label
for i in kk:
    label.append(i[1])

FinalZS['label']=label
FinalZS=FinalZS.sort_values(by=['年'],ascending=True)
```
